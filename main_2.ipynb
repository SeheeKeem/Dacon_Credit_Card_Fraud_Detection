# 3-1-(1). KMeans 학습

```python
from sklearn.preprocessing import scale
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
train_copy = train_scaled.copy()
```

```python
train_copy
```

```python
kmeans = KMeans(n_clusters = 2, init = 'k-means++', max_iter = 300, random_state = 0)
kmeans.fit(train_copy)
kmean_train = kmeans.transform(train_copy)
```

```python
kmean_train
```

```python
pd.DataFrame(kmeans.labels_).value_counts()
```

```python
print(kmeans.labels_)
```

```python
# 축
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# 점들을 데이터 프레임 형테로 저장
kmean_train_df = pd.DataFrame(kmean_train, columns=['X', 'Y'])
display(kmean_train_df.head())

# 점 데이터를 X-Y grid에 시각화하기
ax.scatter(kmean_train_df['X'], kmean_train_df['Y'], c='black', label='train data')


ax.set_xlabel('X') #x축 라벨
ax.set_ylabel('Y')#y축 라벨
ax.legend()
ax.grid()
```

```python
label_0 = kmean_train[kmeans.labels_ == 0]
label_1 = kmean_train[kmeans.labels_ == 1]
```

```python
label_0_df = pd.DataFrame(label_0, columns=['X', 'Y'])
label_1_df = pd.DataFrame(label_1, columns=['X', 'Y'])
```

```python
plt.scatter(label_0_df['X'], label_0_df['Y'], color = 'green', label = "0")
plt.scatter(label_1_df['X'], label_1_df['Y'], color = 'red', label = "1")
```

# 3-1-(2) KMeans 검증

##### Class 컬럼은 실제 validation의 class값, Cluster 컬럼은 validation의 예측 class값

```python
val_copy = val_df.copy().drop(columns = ['ID', 'Class'], axis = 1)
```

```python
kmean_val = kmeans.transform(val_copy)
```

```python
kmean_val
```

```python
kmean_val_df = pd.DataFrame(kmean_val, columns = ['X', 'Y'])
```

```python
kmean_val_df['Class'] = val_df['Class']
```

```python
kmean_val_df
```

```python
class_0 = kmean_val_df[kmean_val_df.Class == 0]
class_1 = kmean_val_df[kmean_val_df.Class == 1]
```

```python
plt.scatter(class_0['X'], class_0['Y'], color = 'green', label = "0")
plt.scatter(class_1['X'], class_1['Y'], color = 'red', label = "1")
```

```python
cluster = kmeans.predict(val_copy)
cluster
```

```python
pd.DataFrame(cluster).value_counts()
```

```python
kmean_val_df['Cluster'] = pd.DataFrame(cluster)
```

```python
kmean_val_df
```

```python
cluster_0 = kmean_val_df[kmean_val_df.Cluster == 0]
cluster_1 = kmean_val_df[kmean_val_df.Cluster == 1]
```

```python
plt.scatter(cluster_0['X'], cluster_0['Y'], color = 'green', label = "0")
plt.scatter(cluster_1['X'], cluster_1['Y'], color = 'red', label = "1")
```

```python
kmean_result = kmean_val_df.groupby(['Class', 'Cluster']).count()
print(kmean_result)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
print(confusion_matrix(val_df['Class'], cluster))
print(accuracy_score(val_df['Class'], cluster))
print(classification_report(val_df['Class'], cluster))
```
